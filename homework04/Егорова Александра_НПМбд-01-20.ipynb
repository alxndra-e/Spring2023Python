Аналитическое решение интеграла №4$: \frac{\pi}{4} \approx 0.78539816$

Аналитическое решение интеграла №9$: \frac{\pi}{2^\frac{3}{2}} \approx 1.11072073$

```python
import math


def f4(x):
    return 1 / (x * (1 + np.log(x) ** 2))

def rectangle(f4, a, b, h):
    x = np.linspace(a, b - h, int((b - a) / h))
    y = h * f4(x)
    return np.sum(y)

def trapezoid(f4, a, b, h):
    x = np.linspace(a, b, int((b - a) / h) + 1)
    y = f4(x[0]) + 2 * np.sum(f4(x[1:-1])) + f4(x[-1])
    return y * h / 2

def simpson(f4, a, b, h):
    x = np.linspace(a, b, int((b - a) / h) + 1)
    y = f4(x[0]) + 4 * np.sum(f4(x[1:-1:2])) + 2 * np.sum(f4(x[2:-1:2])) + f4(x[-1])
    return y * h / 3

print('Метод прямоугольников для №4:', rectangle(f4, 1, math.e, 0.01))
print('Метод трапеций для №4:', trapezoid(f4, 1, math.e, 0.001))
print('Метод Симпсона для №4:', simpson(f4, 1, math.e, 0.0001))
```
