## Exercise 1  
1. Создайте график $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$, используя 250 точек. Не забудьте пометить оси!


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import scipy.special
```

```python
x = np.linspace(-1, 1, 250)
y = (x**2*np.sin(1/x**2))+x

fig, ax = plt.subplots(figsize=(8,5))
plt.plot(x, y, color = 'purple')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
```

2. Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)

```python
x = np.linspace(5,25)
y = abs(1-((1/(1+x**2))/(1/x**2)))

fig, ax = plt.subplots(figsize=(8,5))
ax.semilogy(x, y, color = 'black', linestyle='--')
ax.set_xlabel('Ось x')
ax.set_ylabel('Ось y')
plt.grid()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6*np.pi, 6*np.pi, 250)
y = np.sin(x)/x
fig, ax = plt.subplots(figsize = (15, 5))


ax.plot(x, y, 'r', marker='$\u03C0$', markevery=10)
ax.set_xlabel('Ось x')
ax.set_ylabel('Ось y')
ax.grid()

labels = [(str(-12+i) + 'π/2' if i != 12 else '0') for i in range(25)]
plt.xticks(np.arange(-6*np.pi, 6*np.pi+np.pi/2, np.pi/2), labels)
ax.legend()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.

```python
xx = np.linspace(-np.pi, np.pi, 150)
yy = np.linspace(-np.pi, np.pi, 150)
X, Y = np.meshgrid(xx, yy)
Z = np.sin(X) * np.sin(Y)
fig, ax = plt.subplots(figsize=(6, 5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, cmap='cubehelix')

ax.set_xlabel('Ось x')
ax.set_ylabel('Ось y')
```

2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
xx = np.linspace(-3, 3, 150) 
yy = np.linspace(-3, 3, 150)
X, Y = np.meshgrid(xx, yy)
Z = np.exp(- np.power(X, 2) - np.power(Y, 2))
fig, ax = plt.subplots(figsize=(6, 5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = ax.pcolor(X, Y, Z, norm=norm, cmap='magma', shading='auto')


ax.set_xlabel('Ось x')
ax.set_ylabel('Ось y')
```
